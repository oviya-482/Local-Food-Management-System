```python
import requests 

from tqdm.notebook import tqdm, trange 

from scrapy import Selector 

import pandas as pd
```

```python
url = "https://www.shopsy.in/womens-clothing-online"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)

print(response.status_code)
```

```python
sel = Selector(text=response.text)
```

```python
name = sel.css("div > div.css-175oi2r.r-1habvwh.r-eqz5dr.r-1h0z5md.r-cnw61z.r-1ss6j8a.r-1hvjb8t > div.css-146c3p1.r-dnmrzs.r-1udh08x.r-1udbk01.r-3s2u2q.r-1iln25a > span::text").getall()
```

```python
name
```

```python
offer = sel.css("div.css-175oi2r.r-1habvwh.r-eqz5dr.r-1h0z5md.r-cnw61z.r-1ss6j8a.r-1hvjb8t > div:nth-child(2) > div:nth-child(1) > div:nth-child(1) > div::text").getall()
```

```python
offer
```

```python
actual_price = sel.css("div.css-146c3p1.r-cqee49.r-1vgyyaa.r-1rsjblm.r-13hce6t::text").getall()
```

```python
actual_price
```

```python
ratings = sel.css("div.css-175oi2r.r-1awozwy.r-13awgt0.r-18u37iz.r-1h0z5md.r-f1odvy > div > div.css-175oi2r.r-1awozwy.r-cdmcib.r-jwli3a.r-6koalj.r-18u37iz.r-1enofrn.r-1ss6j8a > div::text").getall()
```

```python
ratings
```

```python
min_len = min(len(name), len(offer), len(actual_price), len(ratings))
min_len
```

```python
name = name[:min_len]
actual_price = actual_price[:min_len]
offer = offer[:min_len]
ratings = ratings[:min_len]
```

```python
df = pd.DataFrame({'name':name,
                   'actual_price':actual_price,
                   'offer':offer,
                   'ratings':ratings})
```
