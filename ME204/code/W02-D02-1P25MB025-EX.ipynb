```python
import pandas as pd
from tqdm.notebook import tqdm
import requests
from scrapy import Selector
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
```

```python
url = 'https://en.wikipedia.org/wiki/Talk:African_National_Congress'
response = requests.get(url, headers=headers)
response.status_code
```

```python
sel = Selector(text= response.text)
```

```python
title = sel.css('span.mw-page-title-main::text').get()
```

```python
print(title)
```

```python
content = sel.css('.banner-shell-header b::text').get()
```

```python
content
```

```python
cat = sel.css('.wpb-header-combined a::text').getall()
```

```python
cat
```

```python
cat_url = sel.css('.wpb-header-combined a::attr(href)').getall()
```

```python
cat_url
```

```python
df = pd.DataFrame({'name':title, "url":url, 'content_assessment':content, 'categories':[cat], 'categories_url':[cat_url]})
```

```python
df
```

```python
cat_dict = {cat[i]: cat_url[i] for i in range(len(cat))}
```

```python
cat_dict
```

```python
df = pd.DataFrame({
    'name' : title,
    'url' : url,
    'content_class' : content,
    'categories_and_urls' : [cat_dict]
})
```

```python
df['categories_and_urls']= df['categories_and_urls'].apply(lambda x: list(x.items()))
```
