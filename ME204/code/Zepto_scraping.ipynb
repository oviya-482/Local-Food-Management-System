```python
import requests                         # for sending HTTP requests

from tqdm.notebook import tqdm, trange  # for progress bars

from scrapy import Selector             # for parsing HTML content
```

```python
url = "https://www.zepto.com/?srsltid=AfmBOorZ3M-yGpUiz1PDmuqaGltonRG5K673_NSCwwQILNWTt_LgZden"
```

```python
headers = {
    'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
response= requests.get(url, headers=headers)
```

```python
response
```

```python
sel = Selector(text=response.text)
```

```python
header = sel.css('oD_gP::text').get()
```

```python
header
```

```python
product_name = sel.css('div[data-slot-id= ProductName] span ::text').getall()
```

```python
product_name
```

```python
variant = sel.css('div.cbfVF3::text').getall()
```

```python
variant
```

```python
rating = sel.css('.cPdMhy::text').getall()
```

```python
rating
```

```python
review_count = sel.css('.cuNaP7::text').getall()
```

```python
review_count
```

```python
# Clean list: remove blanks, strip spaces, remove parentheses, convert 'k' to integers
cleaned = []
for x in review_count:
    val = x.strip("() ").strip()   # remove parentheses + spaces
    if val:  # only proceed if not empty
        if val.endswith('k'):
            num = float(val[:-1]) * 1000
            cleaned.append(int(num))
        else:
            cleaned.append(val)

print(cleaned)
```

```python
packing = sel.css("div[data-slot-id= PackSize] span ::text").getall()
```

```python
packing
```

```python
image_url = sel.css('img.c2ahfT::attr(src)').getall()
```

```python
image_url
```

```python
offer_price = sel.css('div[data-slot-id= EdlpPrice] span ::text').getall()
```

```python
offer_price
```

```python
# Combine rupee symbol with the number
offer_price_cleaned = []
for i in range(0, len(offer_price), 2):  # step by 2
    symbol = offer_price[i].strip()
    num = offer_price[i+1].strip()
    offer_price_cleaned.append(symbol + num)

print(offer_price_cleaned)
```

```python
actual_price = sel.css('.cx3iWL::text').getall()
```

```python
actual_price
```

```python
# Combine rupee symbol with the number
cost_price = []
for i in range(0, len(actual_price), 2):  # step by 2
    symbol = actual_price[i].strip()
    num = actual_price[i+1].strip()
    cost_price.append(symbol + num)

print(cost_price)
```

```python
discount = sel.css('div.c0aYst.cTlV3p > div.cYCsFo > span:nth-child(1)::text').getall()
```

```python
discount
```

```python
data = []
```

```python
# Store data in list
for i in range(len(product_name)):
     data.append({
                "Product Name": product_name[i] if i < len(product_name) else None,
                "Variant": variant[i] if i < len(variant) else None,
                "Cost price": cost_price[i] if i < len(cost_price) else None,
                "Discount": discount[i] if i < len(discount) else None,
                "Review Count": cleaned [i] if i < len(cleaned) else None,
                "Rating": rating[i] if i < len(rating) else None,
                "Packing": packing[i] if i < len(packing) else None,
                "Offer Price": offer_price_cleaned[i] if i < len(offer_price_cleaned) else None,
                "Image URL": image_url[i] if i < len(image_url) else None,
            })
```

```python
min_length=min(len(product_name), len(variant), len(rating), len(cleaned), len(packing), len(image_url), len(offer_price_cleaned), len(cost_price), len(discount))
```

```python
min_length
```

```python
len(product_name), len(variant), len(rating), len(cleaned), len(packing), len(image_url), len(offer_price_cleaned), len(cost_price), len(discount)
```

```python
product_name = product_name[:min_length]
variant = variant[:min_length] 
rating = rating[:min_length]
cleaned = cleaned[:min_length]
packing = packing[:min_length]
image_url = image_url[:min_length]  
offer_price_cleaned = offer_price_cleaned[:min_length]
cost_price = cost_price[:min_length]
discount = discount[:min_length]
```

```python
import pandas as pd
df = pd.DataFrame({
    'Product Name': product_name,
    'Variant': variant,
    'Rating': rating,
    'Review Count': cleaned,
    'Packing': packing,
    'Image URL': image_url,
    'Offer Price': offer_price_cleaned,
    'Cost price': cost_price,
    'Discount': discount,
})
```
