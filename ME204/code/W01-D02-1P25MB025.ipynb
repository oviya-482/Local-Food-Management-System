Take a look at the columns data-product-name and product-name. Write code that establishes, without a doubt, whether these two columns are identical or not. If they are identical, drop one of the columns. If they are not, how would you modify the column names to make them consistent?

```python
print(f"Are the columns identical? {df['product-name'].equals(df['data-product-name'])}")

if df['product-name'].equals(df['data-product-name']):
    print("The columns are identical. Dropping the 'data-product-name' column as it is redundant.")
    df = df.drop(columns=['data-product-name'])
else:
    print("The columns are not identical. Consider renaming the columns.")
```

Change the column names to more concise and meaningful names. For example, data-product-id can be renamed to the much simpler id, product-name to name, and so on. Feel free to make your own choices, just be consistent.

```python
df = df.rename(columns={"product-name":"name",
                                      "data-product-id":"id",
                                      "item-price":"price"})
```

Remove all the £ symbols from the price column.

```python
df.loc[df["price"].str.contains("p"), "price"] = df["price"].apply(lambda x: "0."+ str.replace(x,"p",''))
```

```python
df.loc[df["price"].str.contains("£"), "price"]= df["price"].str.replace("£","")
```

```python
df['price'] = df['price'].astype('float')
```

Remove all the £ symbols from the price column.

```python
item_price_list = df['price'].tolist()
new_list = []
for element in item_price_list:
    element_str = str(element)
    if '£' in element_str:
        new_element = element_str.replace("£", "")
    elif "p" in element_str:
        new_element = element_str.replace("p", "")
        new_element = "0." + new_element
    else:
        new_element = element_str
    new_list.append(new_element)
df["item_price_fixed"] = new_list
df["item_price_fixed"] = df["item_price_fixed"].astype(float)
```
