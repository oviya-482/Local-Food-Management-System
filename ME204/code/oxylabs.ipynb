```python
import requests
from parsel import Selector

data = []
links = ["https://sandbox.oxylabs.io/products"]

for link in links:
    try:

        # Make a request to each URL
        response = requests.get(link, headers={"User-Agent": "Mozilla/5.0"})
        response.raise_for_status()

        # Parse the HTML response
        sel = Selector(text=response.text)

        # Extract data using CSS selectors
        g_name = sel.css('h4::text').getall()
        genre = sel.css(' p.category.css-8fdgzc.eag3qlw9 > span:nth-child(1)::text').getall()
        price = sel.css('div.price-wrapper.css-li4v8k.eag3qlw4::text').getall()
        

        # Store data in list
        for i in range(len(g_name)):
            data.append({
                "Game Name": g_name[i] if i < len(g_name) else None,
                "Price": price[i].strip() if i < len(price) else None,
                "Product links": "https://sandbox.oxylabs.io" + link if link else None,
                "Genre": genre[i] if i < len(genre) else None,
            })

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {link}: {e}")

# Print results
for item in data:
    print(item)
```
