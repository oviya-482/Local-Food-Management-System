```python
import requests                         # for sending HTTP requests

import pandas as pd                       # for data manipulation and analysis

from tqdm.notebook import tqdm, trange  # for progress bars

from scrapy import Selector             # for parsing HTML content
```

```python
url = "https://books.toscrape.com/"
```

```python
response = requests.get(url)
sel = Selector(text=response.text)
```

```python
response
```

```python
sel
```

```python
heading = sel.css("h1::text").get()
```

```python
heading
```

```python
price = sel.css(".price_color::text").getall()
```

```python
price
```

```python
stock = sel.css(".instock.availability::text").getall() 
```

```python
stock
```

```python
stock = [s.strip() for s in stock if s.strip()]
```

```python
stock
```

```python
rating = sel.css("p.star-rating::attr(class)").getall()
```

```python
rating
```

```python
clean_rating = [r.split()[-1] for r in rating]
```

```python
clean_rating
```

```python
book_name = sel.css("h3 > a::text").getall()
```

```python
book_name
```

```python
book_url = sel.css("div.image_container a::attr(href)").getall()
```

```python
book_url
```

```python
len(stock), len(price), len(clean_rating), len(book_name), len(book_url)
```

```python
df = pd.DataFrame({
    "Name": book_name,
    "Price": price,
    "Stock Availability": stock,
    "Rating": clean_rating,
    "URL": book_url
})
```

```python
df
```

```python
df.to_csv("books.csv", index=False)
```
