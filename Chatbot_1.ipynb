# Financial Data Chatbot
This notebook demonstrates a simple chatbot that can answer predefined financial queries using the uploaded dataset.

```python
import pandas as pd

# Load the uploaded financial data
data = {
    'Company': ['Apple', 'Microsoft', 'Tesla'],
    'Year': [2026, 2026, 2025],
    'Revenue': [69926000000, 69926000000, 69926000000],
    'NetIncome': [42097000000, 27747000000, 34922000000],
    'TotalAssets': [359241000000, 359241000000, 359241000000],
    'TotalLiabilities': [291107000000, 274427000000, 54941000000],
    'OperatingCashFlow': [53925000000, 35758000000, 14747000000]
}

df = pd.DataFrame(data)
df
```

### Define the logic behind the chatbot

```python
# Define chatbot logic
def simple_chatbot(user_query, company):
    row = df[df['Company'] == company].iloc[0]
    if user_query == "What is the total revenue?":
        return f"The total revenue of {company} is {row['Revenue']:,}."
    elif user_query == "How has net income changed over the last year?":
        return f"{company}'s net income is {row['NetIncome']:,} in {row['Year']}."
    elif user_query == "What are the total assets?":
        return f"{company} has total assets worth {row['TotalAssets']:,}."
    elif user_query == "What are the total liabilities?":
        return f"{company} has total liabilities of {row['TotalLiabilities']:,}."
    elif user_query == "What is the operating cash flow?":
        return f"{company}'s operating cash flow is {row['OperatingCashFlow']:,}."
    else:
        return "Sorry, I can only provide information on predefined queries."

# Example usage
print(simple_chatbot("What is the total revenue?", "Apple"))
print(simple_chatbot("What are the total liabilities?", "Tesla"))
```

### Documentation
- **Predefined queries:** Revenue, Net Income, Total Assets, Total Liabilities, Operating Cash Flow.
- **Logic:** Uses simple `if-else` statements to match queries.
- **Limitations:** Only works for predefined queries and companies in the dataset.
